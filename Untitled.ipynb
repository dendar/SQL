# Connecting to a Database

- Connnect to SQl database
- Retrieve all information from SQL table
- Retrieve a subset of records from a table using a wher clause 



```python
import pandas as pd
```

```python
# connect darabase and create cursor
import sqlite3 
conn = sqlite3.connect('data.sqlite')
cur = conn.cursor()
```

```python
cur.execute("""SELECT * FROM employees LIMIT 5;""")
```

```python
# The execute command itself only returns the cursor object.
# To see the results, you must use the fetchall method afterwards.
cur.fetchall()
```

```python
# Wrapping Results Into Pandas DataFrames
cur.execute(""" SELECT * FROM employees Limit 4;""")
df =pd.DataFrame(cur.fetchall())
df.head()
```

```python
cur.description
```

```python
# We dont have the columns names. We can access the column  names by calling "cur.description"

cur.execute("""SELECT *  FROM employees LIMIT 5; """)
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df.head()
```

# Where Clause

The WHERE clause filters query results by some condition also combine multiple conditions


```python
# select customers from specific city
cur.execute("""SELECT * FROM  customers WHERE city == 'Boston'; """)
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df.head()
```

```python
# Select multiple Cities
cur.execute("""SELECT * FROM customers WHERE city = 'Boston' OR city = 'Madrid';""")
df = pd.DataFrame(cur.fetchall())
df.columns =[x[0] for x in cur.description]
df
```
